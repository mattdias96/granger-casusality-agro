```python
import pandas as pd
import numpy as np
import os
```

```python
big_df = pd.DataFrame()
for filename in os.listdir(os.getcwd()):
    if 'csv' in filename:
        with open(os.path.join(os.getcwd(), filename), 'r') as f:
            temp_df = pd.read_csv(filename).iloc[:25, :] # Remove max, min, median
            temp_df['State'] = filename[17:-4] # Add state's name
            big_df = pd.concat([big_df, temp_df])
            
```
